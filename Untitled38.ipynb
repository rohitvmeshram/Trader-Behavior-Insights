# Step 3: Data Preprocessing

## 3.1 Process Fear & Greed Index

```python
# ## Step 3: Data Preprocessing

# 3.1 Process Fear & Greed Index
# Convert 'date' to datetime
fear_greed_df['date'] = pd.to_datetime(fear_greed_df['date'], format='%Y-%m-%d')

# Drop unnecessary columns if needed (e.g., timestamp if date is sufficient)
fear_greed_df = fear_greed_df[['date', 'value', 'classification']]

# Map classifications to numerical for analysis (optional)
classification_map = {
    'Extreme Fear': 1,
    'Fear': 2,
    'Neutral': 3,
    'Greed': 4,
    'Extreme Greed': 5
}
fear_greed_df['class_num'] = fear_greed_df['classification'].map(classification_map)
```

## 3.2 Process Historical Trading Data

```python
# 3.2 Process Historical Trading Data
# Parse 'Timestamp IST' (format: DD-MM-YYYY HH:MM)
def parse_timestamp_ist(ts):
    try:
        return datetime.strptime(ts, '%d-%m-%Y %H:%M')
    except:
        return pd.NaT

historical_df['timestamp_parsed'] = historical_df['Timestamp IST'].apply(parse_timestamp_ist)

# Extract date for merging
historical_df['date'] = historical_df['timestamp_parsed'].dt.date
historical_df['date'] = pd.to_datetime(historical_df['date'])  # Convert to datetime for merging

# Handle 'Closed PnL' - ensure it's numeric
historical_df['Closed PnL'] = pd.to_numeric(historical_df['Closed PnL'], errors='coerce').fillna(0)

# Filter relevant columns
historical_df = historical_df[['Account', 'Coin', 'Execution Price', 'Size USD', 'Side', 'date', 'Closed PnL', 'Fee']]

# Calculate Net PnL (PnL minus Fee)
historical_df['Net PnL'] = historical_df['Closed PnL'] - historical_df['Fee']
```

## 3.3 Aggregate Daily Performance

```python
# 3.3 Aggregate Daily Performance
# Group by date and account (assuming multiple accounts, but focus on main one if needed)
daily_performance = historical_df.groupby(['date', 'Account']).agg({
    'Net PnL': 'sum',
    'Size USD': 'sum',  # Total trade volume
    'Closed PnL': 'count'  # Number of trades
}).reset_index()

daily_performance.rename(columns={'Closed PnL': 'Trade Count'}, inplace=True)

# ## Step 4: Merge Datasets
merged_df = pd.merge(daily_performance, fear_greed_df, on='date', how='inner')

# Display merged data sample
print("\nMerged Data Sample:")
display(merged_df.head())
```

# Step 4: Merge Datasets

```python
merged_df = pd.merge(daily_performance, fear_greed_df, on='date', how='inner')

# Display merged data sample
print("\nMerged Data Sample:")
display(merged_df.head())
```

# Step 5: Analysis

## 5.1 Correlation Analysis

```python
# 5.1 Correlation Analysis
# Correlation between Net PnL and Fear/Greed Value
correlation_pnl_value = merged_df['Net PnL'].corr(merged_df['value'])
correlation_pnl_class = merged_df['Net PnL'].corr(merged_df['class_num'])

print(f"Correlation between Net PnL and Fear/Greed Index Value: {correlation_pnl_value:.2f}")
print(f"Correlation between Net PnL and Sentiment Classification (Numeric): {correlation_pnl_class:.2f}")
```

## 5.2 Grouped Analysis: Average PnL by Sentiment Classification

```python
# 5.2 Grouped Analysis: Average PnL by Sentiment Classification
grouped_sentiment = merged_df.groupby('classification').agg({
    'Net PnL': 'mean',
    'Trade Count': 'mean',
    'Size USD': 'mean'
}).reset_index()

print("\nAverage Metrics by Sentiment Classification:")
display(grouped_sentiment)
```
